```python
# k-fold   
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as pt
```

```python
!git clone https://github.com/taehojo/data.git
df = pd.read_csv("./data/sonar3.csv", header = None)
```

```python
df.head()
```

```python
X = df.iloc[:,0:60]
y = df.iloc[:,60]

#   
k=5
#kfold       
kfold = KFold(n_splits=k,shuffle=True)

#    
acc_score=[]

def model_fn():
  model=Sequential()
  model.add(Dense(24,input_dim = 60,activation='relu'))
  model.add(Dense(10,activation='relu'))
  model.add(Dense(1,activation='sigmoid'))
  return model

# k-fold  k  for 
# split()  k , 
for train_index, test_index in kfold.split(X):# train test 
  X_train,X_test = X.iloc[train_index,:],X.iloc[test_index,:]#   
  y_train,y_test = y.iloc[train_index],y.iloc[test_index]#  1  ,:

  model = model_fn()
  model.compile(loss='binary_crossentropy',optimizer = 'adam',metrics=['accuracy']) #   binary_crossentropy  adam 
  history=model.fit(X_train,y_train,epochs=200,batch_size=10,verbose=0)# verbose   0 X_train,y_train   200  batch_size 10
  # 
  accuracy = model.evaluate(X_test,y_test)[1]
  acc_score.append(accuracy)

  #k    
  avg_acc_score=sum(acc_score)/k
  # 
  print(':',acc_score)
  print(' ',avg_acc_score)

```

```python
from sklearn.model_selection import train_test_split
df = pd.read_csv('./data/wine.csv',header = None)

X = df.iloc[:,0:12]
y = df.iloc[:,12]

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,shuffle = True)
# 
model = Sequential()
model.add(Dense(30,input_dim = 12, activation='relu'))
model.add(Dense(12,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
model.summary()


model.compile(loss='binary_crossentropy',optimizer = 'adam', metrics = ['accuracy'])

history = model.fit(X_train,y_train,epochs = 50, batch_size=500,validation_split=0.25) #  20  60  20

score = model.evaluate(X_test,y_test)
print(score[1])
```

```python
#   
#      
modelpath = "./data/model/all/{epoch:20d}-{val_accuracy:.4f}.hdf5"
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint
checkpointer = ModelCheckpoint(filepath=modelpath,verbose=1)
```

```python
# 
history = model.fit(X_train,y_train,epochs = 50,batch_size = 500, validation_split = 0.25,verbose=0,callbacks = [checkpointer])
# test 
score = model.evaluate(X_test,y_test)
print('Test accuracy:',score[1])
```

```python
#    epoch 
best_epoch = max(range(1, len(history.history['accuracy']) + 1), key=lambda x: history.history['accuracy'][x - 1])
best_accuracy = history.history['accuracy'][best_epoch - 1]

print('Best epoch:', best_epoch)
print('Best accuracy:', best_accuracy)
```
